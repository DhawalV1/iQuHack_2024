```python
import numpy as np
import matplotlib.pyplot as plt
x_values_1 = np.linspace(0, 0.25, 100)

# Define the range of x values for the second plot (0.5 to 1)
x_values_2 = np.linspace(0.25, 0.5, 100)
x_values_3 = np.linspace(0.5, 0.75, 100)

# Define the range of x values for the second plot (0.5 to 1)
x_values_4 = np.linspace(0.75, 1 , 100)

x_values_5 = np.linspace(0, 1 , 400)
def f1(x):
    return 0.984375*x 
def f2(x):
    return 0.875*x + 0.03125
def f3(x):
    return 0.6875*x + 0.125

def f4(x):
    return 0.5*x + 0.265625
def f5(x):
    return np.tanh(x)

y_values_1 = f1(x_values_1)

# Compute the y values for the second plot
y_values_2 = f2(x_values_2)

y_values_3 = f3(x_values_3)

y_values_4 = f4(x_values_4)

y_values_5 = f5(x_values_5)

# Create the plot
plt.figure(figsize=(8, 6))

# Plot for the first range
plt.plot(x_values_1, y_values_1, label=r'$0.984375 \times x$ for $0 \leq x \leq 0.5$', color='blue')

# Plot for the second range
plt.plot(x_values_2, y_values_2, label=r'$0.984375 \times x$ for $0.5 \leq x \leq 1$', color='red')

plt.plot(x_values_3, y_values_3, label=r'$0.984375 \times x$ for $0.5 \leq x \leq 1$', color='black')

plt.plot(x_values_4, y_values_4, label=r'$0.984375 \times x$ for $0.5 \leq x \leq 1$', color='green')

plt.plot(x_values_5, y_values_5, label=r'$0.984375 \times x$ for $0.5 \leq x \leq 1$', color='pink')

plt.grid(color='g', linestyle='-', linewidth=0.25)
```
